## Andmete ühendamine ja ettevalmistamine

Selles märkmikus loeme sisse alla laaditud toorandmed (CSV-failid) ja ühendame need üheks terviklikuks andmestikuks.
Kuna algsed andmed on pikas formaadis (iga finantsnäitaja on eraldi real), siis pöörame andmed laia formaati (pivot), kus iga ettevõtte aruanne on ühel real ja finantsnäitajad on tulpades. Lõpuks lisame juurde aruannete metaandmed ja salvestame andmestiku efektiivses `parquet` formaadis edasiseks töötlemiseks.

```python
# Loeme CSV failid sisse ja ühendame aruannete elemendid üheks suureks tabeliks (df)
comp_df = pd.read_csv("raw_data/ettevotja_rekvisiidid__lihtandmed.csv", delimiter=";")
rep_df = pd.read_csv("raw_data/1.aruannete_yldandmed_kuni_31032026.csv", delimiter=";", low_memory=False)

df = pd.concat([
 pd.read_csv("raw_data/4.2019_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 pd.read_csv("raw_data/4.2020_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 pd.read_csv("raw_data/4.2021_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 pd.read_csv("raw_data/4.2022_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 pd.read_csv("raw_data/4.2023_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 pd.read_csv("raw_data/4.2024_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 pd.read_csv("raw_data/4.2025_aruannete_elemendid_kuni_31032026.csv", delimiter=";"),
 ], ignore_index=True)
```

```python
# Vaatame esimesi ridu koondtabelist
df.head()
```

```python
# Tuvastame duplikaadid: osades aruannetes võib olla sama element korduvalt kirjas
dup_keys = (
    df.groupby(["report_id", "elemendi_nimetus"])
    .size()
    .reset_index(name="count")
    .query("count > 1")
    .sort_values(["count", "report_id"], ascending=[False, True])
)

print(dup_keys.head(20))
```

```python
# Eemaldame duplikaadid ja tühjad väärtused võtmeveergudest
# Jätame alles viimase väärtuse korduvatest elementidest
df2 = df.dropna(subset=["report_id", "elemendi_nimetus"])
df2 = df2.reset_index(drop=True)
df2 = df2.drop_duplicates(subset=["report_id", "elemendi_nimetus"], keep="last")
```

```python
# Loome laia formaadis tabeli (pivot), kus iga aruande (report_id) andmed on ühel real ja finantsnäitajad (elemendi_nimetus) tulpades
df_wide = df2.pivot(index="report_id", columns="elemendi_nimetus", values="vaartus").reset_index()
df_wide.head()
```

```python
# Lisame laiale formaadile juurde põhilised metaandmed teisest andmestikust
# Näiteks: ettevõtte registrikood, aruandeaasta, õiguslik vorm, staatus ja auditeerimise info
rep_details = rep_df[["report_id", "registrikood", "aruandeaasta", "õiguslik vorm", "staatus", "kas auditeeritud?", "minimaalne kategooria andmete alusel" ]]

df_wide = df_wide.merge(rep_details, on="report_id")
```

```python
# Ühendame andmed erinevatest andmestikest
rep_details = rep_df[["report_id", "registrikood", "aruandeaasta", "õiguslik vorm", "staatus", "kas auditeeritud?", "minimaalne kategooria andmete alusel" ]]

df_wide = df_wide.merge(rep_details, on="report_id")
```
